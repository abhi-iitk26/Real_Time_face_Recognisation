```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from PIL import Image
import requests
from io import BytesIO
import re

# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Initialize WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get("https://search.pclub.in/")
time.sleep(3)
names = []
rollnos= []
branchs = []
try:
    # Find and click on the Hall button
    print("Finding Hall button...")
    hall_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '/html/body/div/div/div[2]/div/div[3]/div/div/div/div'))
    )
    hall_button.click()
    print("Clicked Hall button.")

    time.sleep(5)  # Wait briefly for dropdown to appear (adjust as needed)

    # Select Hall14 from the dropdown
    print("Selecting Hall 14...")
    select_hall_14 = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '/html/body/div[2]/div[3]/ul/li[9]'))
    )
    select_hall_14.click()
    print("Selected Hall 14.")

    time.sleep(3)  # Wait for page to load (adjust as needed)

    # Fetch and print student details
    for i in range(2, 10):  # Assuming there are 8 students to fetch
        print(f"Fetching details for student {i}...")
        
        name_xpath = f'/html/body/div/div/div[3]/div[2]/div[{i}]/div[2]/p[1]'
        branch_xpath = f'/html/body/div/div/div[3]/div[2]/div[{i}]/div[2]/p[2]'
        rollno_xpath = f'/html/body/div/div/div[3]/div[2]/div[{i}]/div[2]/p[3]'
        photo_xpath = f'/html/body/div/div/div[3]/div[2]/div[{i}]/div[1]'

        name_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, name_xpath))
        )
        name = name_element.text
        names.append(name)
        print("Name:", name)
        time.sleep(2)

        branch_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, branch_xpath))
        )
        branch = branch_element.text
        branchs.append(branch)
        print("Branch:", branch)
        time.sleep(2)

        rollno_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, rollno_xpath))
        )
        rollno = rollno_element.text
        rollnos.append(rollno)
        print("Roll No:", rollno)
        time.sleep(2)

        # Find the photo element and get its style attribute
        print("Finding photo element...")
        photo_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, photo_xpath))
        )
        style_attr = photo_element.get_attribute('style')
        print("Photo element found. Style attribute:", style_attr)

        # Extract URLs from the style attribute
        print("Extracting URLs from style attribute...")
        urls = re.findall(r'url\((.*?)\)', style_attr)
        time.sleep(2)

        if urls:
            # Clean the URLs by removing surrounding quotes
            urls = [url.strip('"').strip('&quot;') for url in urls]

            # Print all extracted URLs
            print("Extracted URLs:", urls)

            # Try each URL until we find a valid image
            for url in urls:
                try:
                    print(f"Trying URL: {url}")
                    response = requests.get(url)
                    img = Image.open(BytesIO(response.content))
                    # Save the image to a file
                    img.save(f'student_photo_{i}.jpg')
                    print("Photo downloaded successfully!")
                    break
                except Exception as e:
                    print(f"Failed to download image from URL: {url}, error: {e}")
        else:
            print("Photo URL is not found.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the WebDriver session
    driver.quit()
    print("WebDriver session closed.")
```
